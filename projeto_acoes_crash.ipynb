```python
#Importando bibliotecas

import yfinance as yf  #Documentação Yfinance--> https://pypi.org/project/yfinance/
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Visualizando o impacto da pandemia na bolsa

Evento também chamado de Coronacrash, mas eu honestamente não curto muito esse nome

```python
b3 = yf.Ticker("B3SA3.SA")
```

```python
hist = b3.history(start="2020-01-01",end="2020-06-01")['Close']
hist.head()
```

```python
hist.plot()
plt.title('Visualizando o impacto na B3', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

Ninguém gosta de trabalhar com incerteza, com isso o índice não só da bolsa brasileira, mas das bolsas ao redor do mundo começaram a capengar até chegar na queda acentuada desde que a OMS classificou a Covid-19 como pandemia

Em contraste, o gráfico abaixo representa o período do ano de 2019

```python
hist = b3.history(start="2019-01-01",end="2019-12-31")['Close']

hist.plot()
plt.title('Índice da B3 no ano de 2019', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

```python
#Atribuindo uma variável com a lista de todas as ações

acoes = ['SUZB3.SA','KLBN4.SA','GOLL4.SA','AZUL4.SA','CVCB3.SA','MELI34.SA','AMER3.SA','MGLU3.SA']
```

```python
#Agora adicionando período de tempo e com apenas a coluna de fechamento das ações no dia

data = yf.download(acoes,start="2019-01-01",end="2022-05-15")['Adj Close']
```

```python
data.tail()
```

# 2.1 Plotando gráficos

Agora podemos visualizar as ações, porém cada ação tem seu preço,ainda precisamos normalizar para deixar as ações na mesma escala

```python
data.plot(figsize=(8,8))
plt.title('Ativos sem normalização', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

### 2.1.2 Normalizando

Ao normalizar, todas as ações partirão do índice do número 1. Facilitando a visualização da valorização ou desvalorização das ações 

```python
normalizado = data/data.iloc[0]

normalizado.head(3)
```

## 2.2 Agora sim os gráficos

```python
#Chamei novamente a "data" par alterar o tempo
data = yf.download(acoes,start="2019-12-01",end="2020-03-20")['Adj Close']

#E novamente normalizado para atribuir o novo valor de data
normalizado = data/data.iloc[0]

normalizado.plot(figsize=(12,6))
plt.title('Ativos normalizados de dezembro de 2019 até o crash', fontsize=20)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

```python
data = yf.download(acoes,start="2020-01-01",end="2020-07-01")['Adj Close']
normalizado = data/data.iloc[0]

normalizado.plot(figsize=(12,6))
plt.title('Ações de 2020 de janeiro até julho ', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

# 3. Queda e ascenção durante a pandemia

No primeiro gráfico podemos ver que todas as ações desvalorizaram em relação à 2019 até o momento inicial, a incerteza levou com ela todo o mercado global e infelizmente o que veio a seguir levaria mais que a bolsa

A partir do crash a dinâmica com que afetou o mercado mudou. As três empresas mais desvalorizadas da lista foram as ligadas à aviação e turismo

# 4. Conclusão

O impacto da pandemia foi diferente para cada setor, eu sei que pode parecer óbvio, mas é legal ter a visualização e explicação na mão. 

As empresas ligadas à compras online cresceram já que o comércio físico estava fechado e muitas pessoas optaram por evitar sair de casa. Com esse aumento de produtos, tivemos o aumento de pacotes e embalagens que precisaram ser produzidos com isso puxando a produção destes pela indústria de papéis e celulose.

Por outro lado, a falta de demanda por turismo ou viagens em geral além de proibições impostas pelo governo fez despencar os papéis de empresas ligadas à viagens. Podemos acompanhar alguns picos que do setor que eram correlatos com notícias de diminuições de restrições e baixa de casos
